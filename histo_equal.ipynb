```python
import cv2
from google.colab.patches import cv2_imshow

# Load the image
img = cv2.imread('/content/Section B.jpg', 0)  # Load as grayscale
print("Image before Histogram Equalization \n")
cv2_imshow(img)

# Perform histogram equalization
equalized_img = cv2.equalizeHist(img)


# Display the equalized image
print("\n Image after Histogram Equalization \n")
cv2_imshow(equalized_img)

# Calculate histograms before and after equalization
hist_before = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_after = cv2.calcHist([equalized_img], [0], None, [256], [0, 256])
```
